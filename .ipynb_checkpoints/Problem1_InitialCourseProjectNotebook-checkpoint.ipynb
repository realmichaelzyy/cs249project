## Data

<b>The data we will be analyzing is a 2006-2013 historical dataset of research awards from the United States government, primarily in the form of NSF and NASA research grants</b> [1]. In analyzing this data, we hope to find patterns in research awards such as popular fields, trending universities, trends in funding to private versus public institutions, and other similar analysis.

This dataset contains a comprehensive set of information for each award: award amount, principal investigators, co-investigators, awardee organization, award start and expiration, award title, abstract, and other information.

There is another dataset that contains a smaller amount of information about each award, but is available farther back in history [2]. NSF awards as far back as 1960 can be obtained, but detailed information such as abstract content is not available for older awards. As we investigate the more recent and abundant dataset, if interesting trends appear that would benefit from further historical analysis we will use this second dataset.

In addition to analysis on the quantitative aspects of the data such as award amounts and awards per university, text analysis of the abstracts may yield interesting results. A simple analysis of the abstracts could identify active and up-and-coming research areas. Deeper mining of the abstracts could investigate if specific wording or characteristics of an abstract relate to how much funding is awarded. Abstract content could be compared with general field popularity to identify how research funding trends compared to overall public perception. (For example, does a spike in DNA sequencing research cause a corresponding peak in the popularity of DNA sequencing keywords as search terms in the following years? Google Trends can serve as an important source for this analysis).

```python
%pylab inline
pylab.xkcd()
pylab.rcParams['figure.figsize'] = 16, 12
```

```python
#Get the data imported from the CSVs and into a dataframe
import pandas as pd
import numpy as np

import re
import glob

# Assemble all of the data into a single DataFrame
pieces = []
#columns = ['name', 'sex', 'births']

yob_filename_re = re.compile(r"exportAwards\-(\d{4})")

yobs = glob.glob("data/*.csv")

for path in yobs:
    year = yob_filename_re.findall(path)[0]
    print(path)
    frame = pd.read_csv(path, low_memory=False)#, names=columns)
    #print(frame)
    
    frame['year'] = int(year)
    pieces.append(frame)

# Concatenate everything into a single DataFrame
nsfdata = pd.concat(pieces, ignore_index=True)
```

### Getting a feel for the data...

```python
#get a feel for the data
print(nsfdata.head(5))
```

### Let's take a look at the data visually:

```python
#plot total NSF funding per year
#==================================

import locale
locale.setlocale(locale.LC_ALL, '')

#used for converting a row that has float "NaN" and String "$78,623,123.34" values to float
def convert(st):
    if (type(st) is str):
        return locale.atof(st.strip("$"))
    else:
        return 0.0

#only get NSF data
payments = nsfdata[nsfdata['Agency'] == 'NSF']
#get these 3 columns
payments = payments[['year', 'Estimated Total Award Amount','Funds Obligated to Date']]
#convert 'Estimated Total Award Amount' to float
payments.ix[:,1] = payments.ix[:,1].apply(convert)
#convert 'Funds Obligated to Date' to float
payments.ix[:,2] = payments.ix[:,2].apply(convert)
#create 'award' column, which is the max of totalawardamount/fundsobligatedtodate.
payments['award'] = payments.ix[:,[1,2]].apply(max,axis=1)

#group by year, sum total awards, then plot dat.
payments.pivot_table('award',rows='year',aggfunc=sum).plot(kind="bar", colormap=matplotlib.cm.get_cmap('summer'))
pylab.title('Total NSF awards per year')
pylab.xlabel('Year')
pylab.ylabel('\$\$ (billions)')

```

### Interesting... funding seems to have stayed pretty constant in the last couple of years, but it looks like 2009 and 2010 saw more money than usual given out by the NSF<br> <br> Let's take a look at a couple of California schools in particular, to see how much NSF money they get

```python
#plot NSF funding per year, for a couple of schools in particular
#==================================

#helper function for checking if a substring is in a string
#takes a substring and a series of strings as input, returns a boolean series
def iselementof(substr, series):
    return series.apply(lambda x: (substr in x) if (type(x) is str) else False)

payments = nsfdata[nsfdata['Agency'] == 'NSF']
payments = payments[['year', 'Estimated Total Award Amount','Funds Obligated to Date','Doing Business As Name']]
payments.ix[:,1] = payments.ix[:,1].apply(convert)
payments.ix[:,2] = payments.ix[:,2].apply(convert)
payments['award'] = payments.ix[:,[1,2]].apply(max,axis=1)


#look at some schools we care about. filter them out
payments = payments[(iselementof('University of California-Los Angeles',payments['Doing Business As Name'])) | \
                    (iselementof('University of California-San Diego',payments['Doing Business As Name'])) | \
                    (iselementof('Stanford University',payments['Doing Business As Name']))]

#group by year and org name. sum 'award'.
schoolData = payments.pivot_table('award',rows=['year'],cols=['Doing Business As Name'],aggfunc=sum)
schoolData.plot(kind="bar", ax=pylab.axes(), colormap=matplotlib.cm.get_cmap('summer'))
pylab.title('Total NSF awards per year, for some California Institutions')
pylab.ylabel('\$\$ (millions)')
```
