## There are a lot of fields with missing values. Let's replace missing values with reasonable substitutes

```python
#fill in values for contract value and WithPHD fields
data.loc[:,'ContractValueBandseenoteA_withnull'] = data.loc[:,'ContractValueBandseenoteA']
data.loc[:,'ContractValueBandseenoteA'] = data.ContractValueBandseenoteA.fillna('-1')

data.loc[:,'GrantCategoryCode_withnull'] = data.loc[:,'GrantCategoryCode']
data.loc[:,'GrantCategoryCode'] = data.GrantCategoryCode.fillna('-1')

data.loc[:,'SponsorCode_withnull'] = data.loc[:,'SponsorCode']
data.loc[:,'SponsorCode'] = data.SponsorCode.fillna('-1')

for i in range(1,14):
    col = "WithPHD" + str(i)
    data.loc[:,col+"_withnull"] = data.loc[:,col]
    data.loc[:,col] = data[col].fillna('No')
    
# fill in values for nonexistant publications    
for i in range(1,14):
    for j in ['NumberofUnsuccessfulGrant','NumberofSuccessfulGrant','Astar','A','B','C']:
        col = j + str(i)
        #print col
        data.loc[:,col + "_withnull"] = data.loc[:,col]
        data.loc[:,col] = data[col].fillna(-1)    
        
data.loc[:,'Startdate'] = data.Startdate.apply(lambda x: datetime.datetime.strptime(x,'%d/%m/%y'))
```

## The amount of the awards is given in bands that range from A-Q representing bins in the range 1-100 million AUD. We need a function that will convert these numbers into their AUD counterparts. We will use the greatest amount in each band.

```python
def maxValue(x):
    if(pandas.isnull(x) or x == '-1'):
        return -1
    x = x.strip()
    if(len(x) > 1):
        print "what? " + str(x) + "length: " + str(len(x))
    if(x == "A"):
        return 50000
    elif(ord(x)-ord("A") < 6):
        return (ord(x)-ord("A"))*100000
    elif(ord(x)-ord("F") < 11):
        return (ord(x)-ord("F"))*1000000
    elif(x == "Q"):
        return 100000000
    else:
        return -1
```

### Therer are a lot of fields that we can combine or mutate to create new features that might be informative. For instance, the research field is a 5-digit number - we want to generate new features that are broader research fields, like the 1st digit and the first 2 digits.

```python
# Add columns for broader RFCD categories (just the first digit and the first 2 digits)
for i in range(1,5):
    oldCol = 'RFCDCode'+str(i)
    newCol = oldCol + 'OneDigit'
    data.loc[:,newCol] = data[oldCol].apply(lambda x: broadRFCD(x,1))
for i in range(1,5):
    oldCol = 'RFCDCode'+str(i)
    newCol = oldCol + 'TwoDigit'
    data.loc[:,newCol] = data[oldCol].apply(lambda x: broadRFCD(x,2))
    
# Add columns that are an aggregation of the broader columns across all of the RFCDs
oneDigitSet = set()
for i in range(1,5):
    colname = 'RFCDCode' + str(i) + 'OneDigit'
    oneDigitSet.update(data[colname].unique())
twoDigitSet = set()
for i in range(1,5):
    colname = 'RFCDCode' + str(i) + 'TwoDigit'
    twoDigitSet.update(data[colname].unique())

containsOneDigitColumns = []
for code in oneDigitSet:
    newColName = 'ContainsRFCDOneDigitCode' + str(code) 
    if code != 0:
            containsOneDigitColumns.append(newColName)
    cols = ['RFCDCode' + str(i) + 'OneDigit' for i in range(1,5)]
    data[newColName] = numpy.logical_or.reduce(data[cols] == code,axis=1)
containsTwoDigitColumns = []
for code in twoDigitSet:
    newColName = 'ContainsRFCDTwoDigitCode' + str(code)
    if code != 0:
        containsTwoDigitColumns.append(newColName)
    cols = ['RFCDCode' + str(i) + 'TwoDigit' for i in range(1,5)]
    data[newColName] = numpy.logical_or.reduce(data[cols] == code,axis=1)

# Just for curiosity, how much of the data covers more than one research field?

print "Proportion of applications that come from more than one broad field " + \
      str(sum(numpy.add.reduce(data[containsOneDigitColumns],axis=1) > 1)/float(len(data)))

print "\nGroup by applications that come from more than one broad research field:"
grouped = data.groupby(numpy.add.reduce(data[containsOneDigitColumns],axis=1) > 1) 
grouped['GrantStatus'].agg({'Total applications': len, 'Success Rate': mean})
```

### Let's generate some features from the investigator roles. Generate features like "ContainsRoleCHIEF_INVESTIGATOR" for each role, which is a boolean indicator to show if any of the investigators in the application fulfill a certain role.

```python
# Add columns that are an aggregation of the roles for each application
roleSet = set()
for i in range(1,14):
    colname = 'Role' + str(i)
    roleSet.update(data[colname].unique())

containsRoleColumns = []
for role in roleSet:
    newColName = 'ContainsRole' + str(role)
    containsRoleColumns.append(newColName)
    cols = ['Role' + str(i) for i in range(1,14)]
    data[newColName] = numpy.logical_or.reduce((data[cols] == role\
                                               if not pandas.isnull(role) else pandas.isnull(data[cols])),axis=1)

# Just for curiosity, how do these new columns look?

print "Proportion of applications that have a chief investigator: " + \
      str(numpy.sum(data['ContainsRoleCHIEF_INVESTIGATOR'],axis=1)/float(len(data)))
    
print "Proportion of applications that have a principle supervisor: " + \
      str(numpy.sum(data['ContainsRolePRINCIPAL_SUPERVISOR'],axis=1)/float(len(data)))
    
print "Proportion of applications that have a student chief investigator: " + \
      str(numpy.sum(data['ContainsRoleSTUD_CHIEF_INVESTIGATOR'],axis=1)/float(len(data)))   
    
grouped = data.groupby(data['ContainsRoleCHIEF_INVESTIGATOR'])
grouped['GrantStatus'].agg({'Total applications': len, 'Success Rate': mean})
```

## <u>Explorations of NSF dataset</u><br> <br> The data we will be exploring in this section consists of NSF awards. Data includes such things as award amount, PI(s), institution(s), abstract data, etc. One thing which the data does NOT have, however, is failed applicants which did not receive awards. Thus, for this section we will be paying attention primarily to exploratory analysis. (Classification or prediction would require "negative" examples -- a.k.a information on what makes a failed application bad)<br> <br> To get started here, let's take a look at the award amounts. We would expect them to be, perhaps, LogNormally distributed (only a few very large awards)<br> <br> We can see below that the lognormal fit is reasonable.

```python
from scipy.stats import norm, lognorm
from numpy import linspace
from pylab import plot,show,hist,figure,title

awards = nsfdata.award.copy()
awards.sort()
awards = awards[awards < 2000000]
awards = awards.apply(lambda x : 1 if x == 0 else x)
#awards = awards.apply(log)
awards = np.array(awards.tolist())

# picking 150 of from a normal distrubution
# with mean 0 and standard deviation 1

param = lognorm.fit(awards) # distribution fitting

# now, param[0] and param[1] are the mean and 
# the standard deviation of the fitted distribution
x = linspace(0,2000000,100)
# fitted distribution
pdf_fitted = lognorm.pdf(x,s=param[0], loc=param[1],scale=param[2])

title('Lognormal fit on hist(award amounts)')
plot(x,pdf_fitted,'r-')
hist(awards,normed=1, bins=20)
show()
```

## To investigate further, let's take a look at a Normal distribution fit to log(award amounts). <br> <br> We can see that the fit isn't nearly as nice when we look at it up up-close. It seems like there are too many awards with small amounts (pushing the mean to the left) to successfully call it lognormal. 

```python
awards = nsfdata.award.copy()
awards.sort()
awards = awards.apply(lambda x : 1 if x == 0 else x)
awards = awards.apply(log)
awards = np.array(awards.tolist())

# picking 150 of from a normal distrubution
# with mean 0 and standard deviation 1

param = norm.fit(awards) # distribution fitting

# now, param[0] and param[1] are the mean and 
# the standard deviation of the fitted distribution
x = linspace(5,20,100)
# fitted distribution
pdf_fitted = norm.pdf(x,loc=param[0],scale=param[1])

title('Normal fit on hist(log(award amounts))')
plot(x,pdf_fitted,'r-')
hist(awards,normed=1, bins=35)
show()
```

## Moving away from Google Trends to define "popularity" in the field, we will appeal to google's "NGrams" feature. Similar to "trends", Google NGrams is restricted to literature searches. This literature includes all books which google has on record, which happen to include proceedings of popular ACM conferences, etc. Thus, we feel that searching for ngrams in this domain produces a relevant metric for the overall "popularity" of the term.<br> <br> Next is the more difficult question of whether or not the NSF is funding this research area.
```python
#Here is a simple NGrams example using the function defined above (runQuery). 
#==========================================
import ngrams

quer = ngrams.runQuery('Machine Learning')
quer.set_index('year',inplace=True)
#using dates for which we have NSF abstract information
quer.plot(xlim=(1983,2008))
pylab.ylabel('percent')
pylab.title('NGrams example')
```

## Next, let's get a sense of how our approach to the popularity of a field using NSF abstracts looks:

```python
def getFreqCount(term, pieces):
    freq = []
    for i in range(len(pieces)):
        abstracts = pieces[i].abstractnarration.dropna()
        count = abstracts.apply(lambda x : 1 if term.lower() in x.lower() else 0).tolist().count(1)
        normct = float(count) / float(len(abstracts))
        freq.append(normct)
    return freq

#simple example, NSF funding for hurricane
freq = getFreqCount('Machine Learning', pieces)
df = pd.DataFrame(freq, columns=['frequency in NSF abstracts'])
df['year'] = range(1983,2009)
df.set_index('year',inplace=True)
df.plot(xlim=(1983,2008))
pylab.title('Machine Learning')
pylab.ylabel('popularity')
```

## Nice. Let's combine those, normalizing along the way so that 1.0 represents the peak in popularity.<br> <br> The results are interesting. The NSF seems to be funding Machine Learning topics in a somewhat similar trend to the popularity of machine leanring in literature.<br> <br> The rise of NSF-funded machine learning, however, seems to lag and be in general less popular than it is in the literature, until around 2007 where the peaks in popularity are similar.

```python
import ngrams

t = 'Machine Learning'
freq = getFreqCount(t, pieces)
quer = ngrams.runQuery(t)
quer.set_index('year',inplace=True)
maxfreq = max(freq)
maxngram = quer.ix[:,0].max()
quer[t] = quer[t].apply(lambda x : x/maxngram)
quer.rename(columns={t : t+" - Google NGrams"}, inplace=True)
quer[t+' - NSF abstracts'] = map(lambda x : 0 if maxfreq == 0 else (x / maxfreq), freq)
quer.plot(xlim=(1983,2008))#, ax=ax[rax,cax])
pylab.title(t)
pylab.ylabel('popularity')
```
