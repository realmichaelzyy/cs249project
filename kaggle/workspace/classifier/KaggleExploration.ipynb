##First, let's read in the data and take a look around



```python
import matplotlib.pyplot as plt
import pandas
import numpy
from numpy import mean
import datetime

%matplotlib inline

data_path = "./data/"

data = pandas.io.parsers.read_csv(data_path + "unimelb_training.csv",low_memory=False)

# Remove dots from the column names (this makes it easier to access the columns in Pandas
data.columns = map(lambda x: 'Astar' + x[3:] if x[0:3] == "A.." else x.replace(".",""), data.columns)
```

####For each application, we have information about the application itself (the sponsoring organization, date of application submission, the grant amount, etc.), the research work (academic field, socioeconomic classification), and about the researchers (role, country of birth, number of publications, previously successful grant applications, etc., for each researcher)

## There are a lot of fields with missing values. Let's replace missing values with reasonable substitutes

```python
#fill in values for contract value and WithPHD fields
data.loc[:,'ContractValueBandseenoteA_withnull'] = data.loc[:,'ContractValueBandseenoteA']
data.loc[:,'ContractValueBandseenoteA'] = data.ContractValueBandseenoteA.fillna('-1')

data.loc[:,'GrantCategoryCode_withnull'] = data.loc[:,'GrantCategoryCode']
data.loc[:,'GrantCategoryCode'] = data.GrantCategoryCode.fillna('-1')

data.loc[:,'SponsorCode_withnull'] = data.loc[:,'SponsorCode']
data.loc[:,'SponsorCode'] = data.SponsorCode.fillna('-1')

for i in range(1,14):
    col = "WithPHD" + str(i)
    data.loc[:,col+"_withnull"] = data.loc[:,col]
    data.loc[:,col] = data[col].fillna('No')
    
# fill in values for nonexistant publications    
for i in range(1,14):
    for j in ['NumberofUnsuccessfulGrant','NumberofSuccessfulGrant','Astar','A','B','C']:
        col = j + str(i)
        #print col
        data.loc[:,col + "_withnull"] = data.loc[:,col]
        data.loc[:,col] = data[col].fillna(-1)    
        
a = data.Startdate.apply(lambda x: datetime.datetime.strptime(x,'%d/%m/%y'))
data.loc[:,'Startdate'] = a
```

## The amount of the awards is given in bands that range from A-Q and $1-1million AUD. We need a function that will convert these numbers into their AUD counterparts. We will use the greatest amount in that band.

```python
def maxValue(x):
    if(pandas.isnull(x) or x == '-1'):
        return -1
    x = x.strip()
    if(len(x) > 1):
        print "what? " + str(x) + "length: " + str(len(x))
    if(x == "A"):
        return 50000
    elif(ord(x)-ord("A") < 6):
        return (ord(x)-ord("A"))*100000
    elif(ord(x)-ord("F") < 11):
        return (ord(x)-ord("F"))*1000000
    elif(x == "Q"):
        return 100000000
    else:
        return -1
```

## Let's do some plots to visualize the numeric data. 

###First we need to refactor some of the fields (for example, sum the number of publications for each researcher in an application to get the total number of publications), and organize these fields into a new dataframe suitable for plotting

```python
from matplotlib.cm import get_cmap
red, blue = '#B2182B', '#2166AC'

# Columns for the successful and unsuccessful grants of each researcher
succGrantCols = ['NumberofSuccessfulGrant' + str(i) + "_withnull" for i in range(1,14)]
unsuccGrantCols = ['NumberofUnsuccessfulGrant' + str(i) + "_withnull" for i in range(1,14)]

# Columns for the total publications and A-grade publications of each researcher
pubcols = [col for col in data.columns if (col.startswith('A') or col.startswith('B') or col.startswith('C')) and col.endswith('withnull') and not col.startswith('Co')]
apubcols = [col for col in data.columns if (col.startswith('A')) and col.endswith('withnull') and not col.startswith('Co')]

plotData = pandas.concat([data.loc[:,['ContractValueBandseenoteA_withnull','GrantStatus']], 
                         pandas.DataFrame({'MaxContractValue':map(lambda x: maxValue(x) if not pandas.isnull(x) else x, data.ContractValueBandseenoteA_withnull),
                                           'totalSuccessfulGrants':numpy.add.reduce(data[succGrantCols].fillna(0),axis=1),
                                           'totalUnsuccessfulGrants':numpy.add.reduce(data[unsuccGrantCols].fillna(0),axis=1),
                                           'totalPubs':numpy.add.reduce(data[pubcols].fillna(0),axis=1),
                                           'totalAPubs': numpy.add.reduce(data[apubcols].fillna(0),axis=1),
                                           'GrantMonth':map(lambda x: x.month, data.Startdate),
                                           'GrantWeek':map(lambda x: x.week, data.Startdate),
                                           'GrantDayofMonth':map(lambda x: x.day, data.Startdate),
                                           'GrantDayofWeek':map(lambda x: x.dayofweek, data.Startdate)})], axis=1)
```

### Let's start by seeing the distribution of grant values and if the value affects whether the grant is accepted or rejected

```python
fig = plt.figure(1, figsize=(16, 8))
ax1 = fig.add_subplot(111)

dropnaData = plotData.dropna()

with plt.xkcd():
    # Cut off very high value data - there's not very much of it, and it makes the plot hard to read
    barData = dropnaData[dropnaData.ContractValueBandseenoteA_withnull.map(
                                lambda x: x != 'Q ' and x != 'P ' and x != 'O ' and x != 'N ' and x != 'M ')]
    
    barData = barData.groupby('ContractValueBandseenoteA_withnull').apply(
                        lambda x: pandas.Series({'Accepted':x.GrantStatus[x.GrantStatus == 1].count(),
                                                 'Rejected': x.GrantStatus[x.GrantStatus == 0].count()}))
    
    plt.bar(range(len(barData)), list(barData.Accepted), color=blue, label='Accepted')
    plt.bar(range(len(barData)), list(barData.Rejected), bottom=list(barData.Accepted), color=red, label='Rejected')
    plt.xticks(numpy.arange(len(barData))+0.5, map(lambda x: "{:,}".format(maxValue(x)), list(barData.index)), rotation='horizontal')
    plt.ylabel("Number of Applications", fontsize=14)
    plt.xlabel("Maximum Grant Value in AUD", fontsize=14)
    plt.legend(loc='upper right', fontsize=14)
    
    plt.show()
```

#### No surprise that the more expensive grant applications are less common. There doesn't seem to be any strong correlation between grant size and chance of being accepted.

### Now let's see the distribution of previous grant applications for researchers, both for applications that are accepted and those that are rejected 

```python
fig = plt.figure(1, figsize=(16, 12))
ax1 = fig.add_subplot(211)

plt.hist(plotData.totalSuccessfulGrants[data.GrantStatus == 1].values, range(0,25,1), color=blue, alpha=0.5, label='Accepted')
#plotData.totalUnsuccessfulGrants[data.GrantStatus == 0]
plt.hist(plotData.totalSuccessfulGrants[data.GrantStatus == 0].values, range(0,25,1), color=red, alpha=0.5, label='Rejected')
plt.xlabel("Number of total successful prior applications", fontsize=14)
plt.legend(fontsize=14)
plt.xlim([0,25])
plt.ylabel("Number of applications",fontsize=14)

ax1 = fig.add_subplot(212)
plt.hist(plotData.totalUnsuccessfulGrants[data.GrantStatus == 1].values, range(0,25,1), color=blue, alpha=0.5, label='Accepted')
#plotData.totalUnsuccessfulGrants[data.GrantStatus == 0]
plt.hist(plotData.totalUnsuccessfulGrants[data.GrantStatus == 0].values, range(0,25,1), color=red, alpha=0.5, label='Rejected')
plt.xlabel("Number of total unsuccessful prior applications", fontsize=14)
plt.xlim([0,25])
plt.ylabel("Number of applications",fontsize=14)
plt.legend(fontsize=14)

plt.show()
```

#### As expected, applications where researchers do not have any prior successful grants do poorly. Conversely, if the researchers do not have any unsuccessful prior applications, the applications tend to do well.

#### These factors seem only to be prescriptive when neither condition exists (either no prior successes or no prior failures). Having a large amount of prior success or failure does not seem to indicate strongly one way or the other

### Now let's do a scatter plot of applications, looking at prior grant history

```python
fig = plt.figure(1, figsize=(12, 10))
ax1 = fig.add_subplot(111)

ax1.scatter(list(numpy.log10(plotData.totalSuccessfulGrants[plotData.GrantStatus == 1])),
                      list(numpy.log10(plotData.totalUnsuccessfulGrants[plotData.GrantStatus == 1])), 
            c='b',cmap=get_cmap('autumn'), alpha = 0.1, s = 196, linewidth=0.0, label = 'Approved')
ax1.scatter(list(numpy.log10(plotData.totalSuccessfulGrants[plotData.GrantStatus == 0])),
                      list(numpy.log10(plotData.totalUnsuccessfulGrants[plotData.GrantStatus == 0])), 
            c='r',cmap=get_cmap('autumn'), alpha = 0.1, s = 200, linewidth=0.0, label = 'Rejected')
#ax1.set_ylim([-1,45])
#ax1.set_xlim([-1,50])
ax1.set_ylim([-0.1,1.7])
ax1.set_xlim([-0.1,1.7])
ax1.set_ylabel('Log of Number of Prior Unsuccessful Grants', fontsize=14)
ax1.set_xlabel('Log of Number of Prior Successful Grants', fontsize=14)
ax1.legend(fontsize=14)
ax1.set_title('Unsuccessful Grants versus Successful Grants',fontsize=14)
ax1.text(1.42,-0.05, "Prior Success",fontsize=14,color='b',alpha=0.7)
ax1.text(-0.08,1.63, "Prior Failure",fontsize=14,color='r',alpha=0.7)
plt.show()
```

#### Again, this confirms our assumptions which is comforting. Dots in the upper left would indicate applications with a prior history of many failures and no successes; dots in the lower right would indicate applications with a prior history of many successes and few failures. We would expect the upper left to have many rejected applications (red), and the bottom right to have successful applications (blue), and that is indeed the trend that we see.
