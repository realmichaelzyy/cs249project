```python
def maxValue(x):
    if(pandas.isnull(x) or x == '-1'):
        return -1
    x = x.strip()
    if(len(x) > 1):
        print "what? " + str(x) + "length: " + str(len(x))
    if(x == "A"):
        return 50000
    elif(ord(x)-ord("A") < 6):
        return (ord(x)-ord("A"))*100000
    elif(ord(x)-ord("F") < 11):
        return (ord(x)-ord("F"))*1000000
    elif(x == "Q"):
        return 100000000
    else:
        return -1
    
grouped = data.groupby(pandas.cut(data.ContractValueBandseenoteA.apply(maxValue),[0,100000,1000000,100000001],labels=["small","med","large"]))
grouped['GrantStatus'].agg({'Total applications': len, 'Success Rate': mean})
```

```python
# the op variable controls whether we are doing logical_or or addition for the reduce operations
def successful(x, pubThresh, grantThresh, op):
    pubcols = []
    pubcols += [col for col in x.index if ((col.startswith('A') or col.startswith('B') or col.startswith('C'))\
        and len(col) < 7 and '14' not in col and '15' not in col)]
    successfulGrantsCols = []
    successfulGrantsCols += [col for col in x.index if 'NumberofSuccess' in col and '14' not in col and '15' not in col]
    unsuccessfulGrantsCols = []
    unsuccessfulGrantsCols += [col for col in x.index if 'NumberofUnsuccess' in col and '14' not in col and '15' not in col]
    if op == 'or':
        return(numpy.logical_or.reduce(x[pubcols] >= pubThresh)[0] and numpy.logical_or.reduce(x[successfulGrantsCols] >= grantThresh)[0])
    else:
        # op == 'add'
        return(numpy.add.reduce(x[pubcols])[0] >= pubThresh and numpy.add.reduce(x[successfulGrantsCols])[0] >= grantThresh)
    
grouped = data.groupby(data.apply(successful,axis=1,args=[1,1,'or']))
grouped['GrantStatus'].agg({'Total applications': len, 'Success Rate': mean})
```

```python
# does having a lot of cumulative experience help?
grouped = data.groupby(data.apply(successful,axis=1,args=[15,6,'add']))
grouped['GrantStatus'].agg({'Total applications': len, 'Success Rate': mean})
```

```python
# does having a superstar help?
grouped = data.groupby(data.apply(successful,axis=1,args=[10,4,'or']))
grouped['GrantStatus'].agg({'Total applications': len, 'Success Rate': mean})
```

```python
# Add columns for broader RFCD categories (just the first digit and the first 2 digits)
for i in range(1,5):
    oldCol = 'RFCDCode'+str(i)
    newCol = oldCol + 'OneDigit'
    data.loc[:,newCol] = data[oldCol].apply(lambda x: broadRFCD(x,1))
for i in range(1,5):
    oldCol = 'RFCDCode'+str(i)
    newCol = oldCol + 'TwoDigit'
    data.loc[:,newCol] = data[oldCol].apply(lambda x: broadRFCD(x,2))
    
# Add columns that are an aggregation of the broader columns across all of the RFCDs
oneDigitSet = set()
for i in range(1,5):
    colname = 'RFCDCode' + str(i) + 'OneDigit'
    oneDigitSet.update(data[colname].unique())
twoDigitSet = set()
for i in range(1,5):
    colname = 'RFCDCode' + str(i) + 'TwoDigit'
    twoDigitSet.update(data[colname].unique())

containsOneDigitColumns = []
for code in oneDigitSet:
    newColName = 'ContainsRFCDOneDigitCode' + str(code) 
    if code != 0:
            containsOneDigitColumns.append(newColName)
    cols = ['RFCDCode' + str(i) + 'OneDigit' for i in range(1,5)]
    data[newColName] = numpy.logical_or.reduce(data[cols] == code,axis=1)
containsTwoDigitColumns = []
for code in twoDigitSet:
    newColName = 'ContainsRFCDTwoDigitCode' + str(code)
    if code != 0:
        containsTwoDigitColumns.append(newColName)
    cols = ['RFCDCode' + str(i) + 'TwoDigit' for i in range(1,5)]
    data[newColName] = numpy.logical_or.reduce(data[cols] == code,axis=1)

# Just for curiosity, how much of the data covers more than one research field?

print "Proportion of applications that come from more than one broad field " + \
      str(sum(numpy.add.reduce(data[containsOneDigitColumns],axis=1) > 1)/float(len(data)))

grouped = data.groupby(numpy.add.reduce(data[containsOneDigitColumns],axis=1) > 1) 
grouped['GrantStatus'].agg({'Total applications': len, 'Success Rate': mean})
```
