```python
#basic imports. 
#=========================================================
%pylab inline
pylab.xkcd()
pylab.rcParams['figure.figsize'] = 16, 9
import pandas as pd
import numpy as np
```

```python
#import data
#=========================================================
import re
import glob
#this list will store a dataframe in each index, corresponding to one year of data.
pieces = []

yob_filename_re = re.compile(r"exportAwards\-(\d{4})")

yobs = glob.glob("data/*.csv")

for path in yobs:
    year = yob_filename_re.findall(path)[0]
    print(path)
    frame = pd.read_csv(path, low_memory=False, parse_dates=True)
    
    frame['year'] = int(year)
    pieces.append(frame)
    print('size: '+str(frame.shape))

# Concatenate everything into a single DataFrame
nsfdata = pd.concat(pieces, ignore_index=True)
nsfdata.rename(columns=lambda x: re.sub("\\W","",x), inplace=True)
nsfdata.set_index(nsfdata.FederalAwardIDNumber, inplace=True)

#convert currency columns to float
#e.g. $123,987.123 to float.
#=========================================================
import locale
locale.setlocale(locale.LC_ALL, '')

def convert(moneyStr):
    if pd.notnull(moneyStr):
        return locale.atof(moneyStr.strip("$"))
    else:
        return 0.0
nsfdata.FundsObligatedtoDate = nsfdata.FundsObligatedtoDate.apply(convert)
nsfdata.EstimatedTotalAwardAmount = nsfdata.EstimatedTotalAwardAmount.apply(convert)
#create a new column, which is the max of these two
nsfdata['award'] = nsfdata[['FundsObligatedtoDate','EstimatedTotalAwardAmount']].apply(max,axis=1)
#convert the dates to pandas "datetime" objects.
nsfdata.AwardDate = nsfdata.AwardDate.apply(pd.to_datetime)
print('total size: '+str(nsfdata.shape))
nsfdata.head(4)
```

## Next, we'll consider the question <i>Does the NSF fund technologies which are (or will become) relevant?</i><br> <br> Obviously this is a difficult question to answer. For now, we'll restrict our analysis to some of the most popular/recent research areas in CS and a couple of other big-name areas (such as clean energy, etc).<br> <br> To define "popularity" in the field, we will appeal to google's "NGrams" feature. Similar to "trends", Google NGrams is restricted to literature searches. This literature includes all books which google has on record, which happen to include proceedings of popular ACM conferences, etc. Thus, we feel that searching for ngrams in this domain produces a relevant metric for the overall "popularity" of the term.<br> <br> Next is the more difficult question of whether or not the NSF is funding this research area. We will determine this as follows. Each award includes the abstract associated with the application. By performing text analysis on these abstracts using keywords in the research area we are investigating, we can get a crude sense of how actively the NSF is funding that area in a given year.<br> <br> To start, let's get a sense of what Google NGrams provides us. This is a simple plot of the term "Machine Learning" from 1983 to 2008:

```python
#Here is a simple NGrams example using the function defined above (runQuery). 
#==========================================
import ngrams

quer = ngrams.runQuery('Machine Learning')
quer.set_index('year',inplace=True)
#using dates for which we have NSF abstract information
quer.plot(xlim=(1983,2008), colormap="gray")
pylab.ylabel('percent')
pylab.title('NGrams example')
```

```python
#working with the XML nsf data which contains abstracts
#note: overriding the "pieces" list created above with different data now.
#=====================================================================
import re
import glob

#from 1976 to 2013
#ABSTRACTS ONLY START APPEARING STARTING AT 1983
dateRange = (1983,2008)

pieces = [];

yob_filename_re = re.compile(r"df\-(\d{4})")

#this is a hardcoded path, change it accordingly. 
yobs = glob.glob("nsf-dataframes/*/*.msg")

for path in yobs:
    #find the year using regex from the filename
    year = int(yob_filename_re.findall(path)[0])
    #only select years within the range defined above
    if year >= dateRange[0] and year <= dateRange[1]:
        print(year)
        print(path)
        df = pd.read_pickle(path)
        df['year'] = year
        #make the column names simpler (remove punctuation, spaces, etc)
        df.rename(columns=lambda x: re.sub("\\W","",x), inplace=True)
        pieces.append(df)
        print('size: '+str(df.shape))
        
#now, "pieces" contains each year.
```

## Next, let's get a sense of how our approach to the popularity of a field using NSF abstracts looks:

```python
def getFreqCount(term, pieces):
    freq = []
    for i in range(len(pieces)):
        abstracts = pieces[i].abstractnarration.dropna()
        count = abstracts.apply(lambda x : 1 if term.lower() in x.lower() else 0).tolist().count(1)
        normct = float(count) / float(len(abstracts))
        freq.append(normct)
    return freq

#simple example, NSF funding for hurricane
freq = getFreqCount('Machine Learning', pieces)
df = pd.DataFrame(freq, columns=['frequency in NSF abstracts'])
df['year'] = range(1983,2009)
df.set_index('year',inplace=True)
df.plot(xlim=(1983,2008), colormap="gray")
pylab.title('Machine Learning')
pylab.ylabel('popularity')
```

## Nice. Let's combine those, normalizing along the way so that 1.0 represents the "peak" in popularity.<br> <br> The results are interesting. The NSF seems to be funding Machine Learning topics in a somewhat similar trend to the popularity of machine leanring in literature.<br> <br> The rise of NSF-funded machine learning, however, seems to lag and be in general less popular than it is in the literature, until around 2007 where the peaks in popularity are similar.

```python
t = 'Machine Learning'
freq = getFreqCount(t, pieces)
quer = ngrams.runQuery(t)
quer.set_index('year',inplace=True)
maxfreq = max(freq)
maxngram = quer.ix[:,0].max()
quer[t] = quer[t].apply(lambda x : x/maxngram)
quer.rename(columns={t : t+" - Google NGrams"}, inplace=True)
quer[t+' - NSF abstracts'] = map(lambda x : 0 if maxfreq == 0 else (x / maxfreq), freq)
quer.plot(xlim=(1983,2008))#, ax=ax[rax,cax])
pylab.title(t)
pylab.ylabel('popularity')
```

## Now, we would like to somehow quantify (at least for the research topics we're looking at) to what extend the NSF funds "relevant" things. That is, does the NSF fund things which are growing similarly in popularity with the literature?<br> <br> Note that there should be a penalty to the NSF for the following two cases: (1) The NSF is failing to fund topics which are extremely popular in the current literature, or (2) the NSF is overly-funding topics which are not popular in the literature.<br> <br> To quantify this, we will appeal to the plots seen above. We will define a "difference" metric which is based on the overall deviation of the NSF funding trends to the literature trends for a given topic.
