```python
#basic stuffs
#=========================================================
%pylab inline
pylab.xkcd()
pylab.rcParams['figure.figsize'] = 16, 9
import pandas as pd
import numpy as np
```

```python
#import data (into var nsfdata)
#=========================================================
import re
import glob
pieces = []

yob_filename_re = re.compile(r"exportAwards\-(\d{4})")

yobs = glob.glob("data/*.csv")

for path in yobs:
    year = yob_filename_re.findall(path)[0]
    print(path)
    frame = pd.read_csv(path, low_memory=False, parse_dates=True)
    
    frame['year'] = int(year)
    pieces.append(frame)
    print('size: '+str(frame.shape))

# Concatenate everything into a single DataFrame
nsfdata = pd.concat(pieces, ignore_index=True)
nsfdata.rename(columns=lambda x: re.sub("\\W","",x), inplace=True)
nsfdata.set_index(nsfdata.FederalAwardIDNumber, inplace=True)

#convert currency columns to float
#=========================================================
import locale
locale.setlocale(locale.LC_ALL, '')

def convert(moneyStr):
    if pd.notnull(moneyStr):
        return locale.atof(moneyStr.strip("$"))
    else:
        return 0.0
nsfdata.FundsObligatedtoDate = nsfdata.FundsObligatedtoDate.apply(convert)
nsfdata.EstimatedTotalAwardAmount = nsfdata.EstimatedTotalAwardAmount.apply(convert)
#create a new column, which is the max of these two
nsfdata['award'] = nsfdata[['FundsObligatedtoDate','EstimatedTotalAwardAmount']].apply(max,axis=1)
#convert the dates to pandas "datetime" objects.
nsfdata.AwardDate = nsfdata.AwardDate.apply(pd.to_datetime)
print('total size: '+str(nsfdata.shape))
nsfdata.head(4)
```
