```python
#Get the data imported from the CSVs and into a dataframe
import pandas as pd
import numpy as np

import re
import glob

# Assemble all of the data into a single DataFrame
pieces = []
#columns = ['name', 'sex', 'births']

yob_filename_re = re.compile(r"exportAwards\-(\d{4})")

yobs = glob.glob("data/*.csv")

for path in yobs:
    year = yob_filename_re.findall(path)[0]
    print(path)
    frame = pd.read_csv(path, low_memory=False)#, names=columns)
    #print(frame)
    
    frame['year'] = int(year)
    pieces.append(frame)

# Concatenate everything into a single DataFrame
nsfdata = pd.concat(pieces, ignore_index=True)
nsfdata.rename(columns=lambda x: re.sub("\\W","",x), inplace=True)
```

### Let's take a look at the data visually:

```python
#plot total NSF funding per year
#==================================

import locale
locale.setlocale(locale.LC_ALL, '')

#used for converting a row that has float "NaN" and String "$78,623,123.34" values to float
def convert(st):
    if (type(st) is str):
        return locale.atof(st.strip("$"))
    else:
        return 0.0

#only get NSF data
payments = nsfdata[nsfdata['Agency'] == 'NSF']
#get these 3 columns
payments = payments[['year', 'EstimatedTotalAwardAmount','FundsObligatedtoDate']]
#convert 'Estimated Total Award Amount' to float
payments.ix[:,1] = payments.ix[:,1].apply(convert)
#convert 'Funds Obligated to Date' to float
payments.ix[:,2] = payments.ix[:,2].apply(convert)
#create 'award' column, which is the max of totalawardamount/fundsobligatedtodate.
payments['award'] = payments.ix[:,[1,2]].apply(max,axis=1)

#group by year, sum total awards, then plot dat.
payments.pivot_table('award',rows='year',aggfunc=sum).plot(kind="bar", colormap=matplotlib.cm.get_cmap('summer'))
pylab.title('Total NSF awards per year')
pylab.xlabel('Year')
pylab.ylabel('\$\$ (billions)')

```

### Interesting... funding seems to have stayed pretty constant in the last couple of years, but it looks like 2009 and 2010 saw more money than usual given out by the NSF<br> <br> Let's take a look at a couple of California schools in particular, to see how much NSF money they get

```python
#plot NSF funding per year, for a couple of schools in particular
#==================================

#helper function for checking if a substring is in a string
#takes a substring and a series of strings as input, returns a boolean series
def iselementof(substr, series):
    return series.apply(lambda x: (substr in x) if (type(x) is str) else False)

payments = nsfdata[nsfdata['Agency'] == 'NSF']
payments = payments[['year', 'EstimatedTotalAwardAmount','FundsObligatedtoDate','DoingBusinessAsName']]
payments.ix[:,1] = payments.ix[:,1].apply(convert)
payments.ix[:,2] = payments.ix[:,2].apply(convert)
payments['award'] = payments.ix[:,[1,2]].apply(max,axis=1)


#look at some schools we care about. filter them out
payments = payments[(iselementof('University of California-Los Angeles',payments['DoingBusinessAsName'])) | \
                    (iselementof('University of California-San Diego',payments['DoingBusinessAsName'])) | \
                    (iselementof('Stanford University',payments['DoingBusinessAsName']))]

#group by year and org name. sum 'award'.
schoolData = payments.pivot_table('award',rows=['year'],cols=['DoingBusinessAsName'],aggfunc=sum)
schoolData.plot(kind="bar", ax=pylab.axes(), colormap=matplotlib.cm.get_cmap('summer'))
pylab.title('Total NSF awards per year, for some California Institutions')
pylab.ylabel('\$\$ (millions)')
```
