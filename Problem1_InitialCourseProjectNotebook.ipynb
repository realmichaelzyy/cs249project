```python
#Get the data imported from the CSVs and into a dataframe
import pandas as pd
import numpy as np

import re
import glob

# Assemble all of the data into a single DataFrame
pieces = []
#columns = ['name', 'sex', 'births']

yob_filename_re = re.compile(r"exportAwards\-(\d{4})")

yobs = glob.glob("data/*.csv")

for path in yobs:
    year = yob_filename_re.findall(path)[0]
    print(path)
    frame = pd.read_csv(path, low_memory=False)#, names=columns)
    #print(frame)
    
    frame['year'] = int(year)
    pieces.append(frame)

# Concatenate everything into a single DataFrame
nsfdata = pd.concat(pieces, ignore_index=True)
```

### Getting a feel for the data...

```python
#get a feel for the data
print(nsfdata.head(5))
```

### Let's take a look at the data visually:

```python
#plot total NSF funding per year
#==================================

import locale
locale.setlocale(locale.LC_ALL, '')

#used for converting a row that has float "NaN" and String "$78,623,123.34" values to float
def convert(st):
    if (type(st) is str):
        return locale.atof(st.strip("$"))
    else:
        return 0.0

payments = nsfdata[nsfdata['Agency'] == 'NSF']
payments = payments[['year', 'Estimated Total Award Amount','Funds Obligated to Date']]
payments.ix[:,1] = payments.ix[:,1].apply(convert)
payments.ix[:,2] = payments.ix[:,2].apply(convert)
payments['award'] = payments.ix[:,[1,2]].apply(max,axis=1)

payments.pivot_table('award',rows='year',aggfunc=sum).plot(kind="bar")
pylab.title('Total NSF awards per year')
pylab.xlabel('Year')
pylab.ylabel('\$\$ (billions)')

```

### Interesting... funding seems to have stayed pretty constant in the last couple of years, but it looks like 2009 and 2010 saw more money than usual given out by the NSF<br> <br> Let's take a look at a couple of California schools in particular, to see how much NSF money they get

```python
#plot NSF funding per year, for a couple of schools in particular
#==================================
payments = nsfdata[nsfdata['Agency'] == 'NSF']
payments = payments[['year', 'Estimated Total Award Amount','Funds Obligated to Date','Awardee']]
payments.ix[:,1] = payments.ix[:,1].apply(convert)
payments.ix[:,2] = payments.ix[:,2].apply(convert)
payments['award'] = payments.ix[:,[1,2]].apply(max,axis=1)

#look at some schools we care about
payments = payments[(payments['Awardee'] == 'UNIVERSITY OF CALIFORNIA, LOS ANGELES') | \
                    (payments['Awardee'] == 'UNIVERSITY OF CALIFORNIA, BERKELEY') | \
                    (payments['Awardee'] == 'UNIVERSITY OF CALIFORNIA, SAN DIEGO') | \
                    (payments['Awardee'] == 'LELAND STANFORD JUNIOR UNIVERSITY, THE')]

schoolData = payments.pivot_table('award',rows=['year','Awardee'],aggfunc=sum).unstack()
schoolData.plot(kind="bar")
pylab.title('Total NSF awards per year, for some California schools')
pylab.xlabel('Year')
pylab.ylabel('\$\$ (millions)')
```
