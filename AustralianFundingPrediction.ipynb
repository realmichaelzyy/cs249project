## There are a lot of fields with missing values. Let's replace missing values with reasonable substitutes

```python
#fill in values for contract value and WithPHD fields
data.loc[:,'ContractValueBandseenoteA_withnull'] = data.loc[:,'ContractValueBandseenoteA']
data.loc[:,'ContractValueBandseenoteA'] = data.ContractValueBandseenoteA.fillna('-1')

data.loc[:,'GrantCategoryCode_withnull'] = data.loc[:,'GrantCategoryCode']
data.loc[:,'GrantCategoryCode'] = data.GrantCategoryCode.fillna('-1')

data.loc[:,'SponsorCode_withnull'] = data.loc[:,'SponsorCode']
data.loc[:,'SponsorCode'] = data.SponsorCode.fillna('-1')

for i in range(1,14):
    col = "WithPHD" + str(i)
    data.loc[:,col+"_withnull"] = data.loc[:,col]
    data.loc[:,col] = data[col].fillna('No')
    
# fill in values for nonexistant publications    
for i in range(1,14):
    for j in ['NumberofUnsuccessfulGrant','NumberofSuccessfulGrant','Astar','A','B','C']:
        col = j + str(i)
        #print col
        data.loc[:,col + "_withnull"] = data.loc[:,col]
        data.loc[:,col] = data[col].fillna(-1)    
        
data.loc[:,'Startdate'] = data.Startdate.apply(lambda x: datetime.datetime.strptime(x,'%d/%m/%y'))
```

## The amount of the awards is given in bands that range from A-Q representing bins in the range 1-100 million AUD. We need a function that will convert these numbers into their AUD counterparts. We will use the greatest amount in each band.

```python
def maxValue(x):
    if(pandas.isnull(x) or x == '-1'):
        return -1
    x = x.strip()
    if(len(x) > 1):
        print "what? " + str(x) + "length: " + str(len(x))
    if(x == "A"):
        return 50000
    elif(ord(x)-ord("A") < 6):
        return (ord(x)-ord("A"))*100000
    elif(ord(x)-ord("F") < 11):
        return (ord(x)-ord("F"))*1000000
    elif(x == "Q"):
        return 100000000
    else:
        return -1
```

```python
from matplotlib.cm import get_cmap
red, blue = '#B2182B', '#2166AC'

# Columns for the successful and unsuccessful grants of each researcher
succGrantCols = ['NumberofSuccessfulGrant' + str(i) + "_withnull" for i in range(1,14)]
unsuccGrantCols = ['NumberofUnsuccessfulGrant' + str(i) + "_withnull" for i in range(1,14)]

# Columns for the total publications and A-grade publications of each researcher
pubcols = [col for col in data.columns if (col.startswith('A') or col.startswith('B') or col.startswith('C')) and col.endswith('withnull') and not col.startswith('Co')]
apubcols = [col for col in data.columns if (col.startswith('A')) and col.endswith('withnull') and not col.startswith('Co')]

plotData = pandas.concat([data.loc[:,['ContractValueBandseenoteA_withnull','GrantStatus']], 
                         pandas.DataFrame({'MaxContractValue':map(lambda x: maxValue(x) if not pandas.isnull(x) else x, data.ContractValueBandseenoteA_withnull),
                                           'totalSuccessfulGrants':numpy.add.reduce(data[succGrantCols].fillna(0),axis=1),
                                           'totalUnsuccessfulGrants':numpy.add.reduce(data[unsuccGrantCols].fillna(0),axis=1),
                                           'totalPubs':numpy.add.reduce(data[pubcols].fillna(0),axis=1),
                                           'totalAPubs': numpy.add.reduce(data[apubcols].fillna(0),axis=1),
                                           'GrantYear':map(lambda x: x.year, data.Startdate),
                                           'GrantMonth':map(lambda x: x.month, data.Startdate),
                                           'GrantWeek':map(lambda x: x.week, data.Startdate),
                                           'GrantDayofMonth':map(lambda x: x.day, data.Startdate),
                                           'GrantDayofWeek':map(lambda x: x.dayofweek, data.Startdate)})], axis=1)
```

### Let's start by seeing the distribution of grant values and if the value affects whether the grant is accepted or rejected

```python
fig = plt.figure(1, figsize=(16, 8))
ax1 = fig.add_subplot(111)

dropnaData = plotData.dropna()

with plt.xkcd():
    # Cut off very high value data - there's not very much of it, and it makes the plot hard to read
    barData = dropnaData[dropnaData.ContractValueBandseenoteA_withnull.map(
                                lambda x: x != 'Q ' and x != 'P ' and x != 'O ' and x != 'N ' and x != 'M ')]
    
    barData = barData.groupby('ContractValueBandseenoteA_withnull').apply(
                        lambda x: pandas.Series({'Accepted':x.GrantStatus[x.GrantStatus == 1].count(),
                                                 'Rejected': x.GrantStatus[x.GrantStatus == 0].count()}))
    
    plt.bar(range(len(barData)), list(barData.Accepted), color=blue, label='Accepted')
    plt.bar(range(len(barData)), list(barData.Rejected), bottom=list(barData.Accepted), color=red, label='Rejected')
    plt.xticks(numpy.arange(len(barData))+0.5, map(lambda x: "{:,}".format(maxValue(x)), list(barData.index)), rotation='horizontal')
    plt.ylabel("Number of Applications", fontsize=14)
    plt.xlabel("Maximum Grant Value in AUD", fontsize=14)
    plt.legend(loc='upper right', fontsize=14)
    
    plt.show()
```

### Research Areas: what fields have the highest success rate?

```python
# research area and grant status
def broadRFCD(x, digits):
    if pandas.isnull(x):
        return -1
    elif x == 0:
        return 0
    return int(x/(pow(10,numpy.floor(numpy.log10(x))+1-digits)))

data['RFCDCode1'].apply(lambda x: broadRFCD(x,2)).nunique()
s = data['RFCDCode1'].apply(lambda x: broadRFCD(x,2))
s.name = 'First 2 Digits of RFCDCode1'
grouped = data.groupby(s)
returndf = grouped.apply(lambda x: 
                         pandas.Series({'Total applications': len(x), 
                                        'Mean Award Amount (1000s AUD)': 
                                            int(x.ContractValueBandseenoteA_withnull.map(lambda x: maxValue(x)).mean()/1000),
                                        'Success Rate': x.GrantStatus.mean()})).sort('Success Rate')
returndf[['Success Rate','Total applications','Mean Award Amount (1000s AUD)']]
```

#### Some interesting tidbits: 
- The fields with the lowest success rate are philosophy and religion, language and culture, and the arts (44, 42, and 41).
- The fields with the highest success rate are economics, commerce and management, and education
- Research fields in the 20s are STEM fields. They seem to be more expensive in general than other fields. 24, the most expensive, is the physical sciences.

## Classification

### Now for the prediction. Do some setup and let's do 4 different classifications using different features. In other words, we want to see how restricting the dataset might affect the classification. For these tests we will be using a random forest classifier.

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, auc
from sklearn import cross_validation, svm
from sklearn.cross_validation import cross_val_score
from sklearn.linear_model import LogisticRegression

randomState = numpy.random.RandomState(0)

# Get the start of the fields in the data that we will use for classificatin
firstColLoc = data.columns.get_loc('RFCDCode1OneDigit')

# Grab lists of some interesting subsets of the fields
dayCols = ['Month', 'DayofWeek', 'DayofMonth','WeekofYear']
grantPubCols = ['ContainsSuccessfulGrant', 'ContainsUnsuccessfulGrant', 'ContainsNoGrantInformation', \
                'ContainsManyUnsuccessfulGrant', 'FirstAuthorSuccessfulGrantRatio', 'SecondAuthorSuccessfulGrantRatio', \
                'ThirdAuthorSuccessfulGrantRatio', 'ContainsHighlySuccessfulGrantAuthor']
encodedCols = ['ContractValueBandseenoteA_enc', 'GrantCategoryCode_enc', 'SponsorCode_enc']
nonPerformanceCols = list(data.columns[data.columns.get_loc('RFCDCode1OneDigit'):data.columns.get_loc('ContainsWithPHDNo')])

# convert boolean fields to numeric
for col in data.iloc[:,firstColLoc:]:
    if(data[col].dtype == bool):
        data.loc[:,col] = le.fit_transform(data.iloc[:,firstColLoc+10])
```

```python
colSets = [dayCols, grantPubCols, encodedCols, nonPerformanceCols] #dayCols + grantPubCols + encodedCols]
colTitles = ['using features from StartDate field',
             'using grant success and publication fields',
             'using contract value, grant category, and sponsor code',
             'using non-performance features']

for i in range(len(colSets)):
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(data.loc[:,colSets[i]], \
                                                                     data['GrantStatus'], test_size=0.4, random_state=randomState)    
    clf = RandomForestClassifier()
    probas_ = clf.fit(X_train,y_train).predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
    roc_auc = auc(fpr, tpr)
    print("Area under the ROC curve : %f" % roc_auc)
    
    fig = plt.figure(1, figsize=(14, 14))
    ax1 = fig.add_subplot(2,2,i+1)
    
    ax1.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
    ax1.plot([0, 1], [0, 1], 'k--')
    ax1.set_xlim([0.0, 1.0])
    ax1.set_ylim([0.0, 1.0])
    ax1.set_xlabel('False Positive Rate')
    ax1.set_ylabel('True Positive Rate')
    ax1.set_title('ROC ' + colTitles[i])
    ax1.legend(loc="lower right")

print('\nROC Curves using different features')
plt.show()
```

#### Of these four feature sets, the non-performance features do the worst. These include the research field, socioeconomic field, investigator countries of birth, investigator roles, etc. The best feature set seems to be the contract value, grant category, and sponsor code - so it looks like various intrinsic features of the grant are good features in general (where the money is coming from, how much money, etc.)

### Now let's use all of the data and try it out on a few different classifier types
### We will try four different classifiers: Logistic Regression, SVM, Random Forest, and AdaBoost

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(data.iloc[:,firstColLoc:], \
                                                                     data['GrantStatus'], test_size=0.4, random_state=randomState)

clfs = [LogisticRegression(),RandomForestClassifier(),svm.SVC(probability=True, random_state=0),AdaBoostClassifier()]
clfNames = ['Logistic Regression','Random Forest','SVM','AdaBoost']

for (i,clf) in enumerate(clfs):
    fittest = clf.fit(X_train,y_train)
    probas_ = fittest.predict_proba(X_test)
    fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
    roc_auc = auc(fpr, tpr)
    print("Area under the ROC curve for %s classifier : %f" % (clfNames[i],roc_auc))
    fig = plt.figure(1, figsize=(14, 14))
    plt.subplot(2,2,i)
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve for ' + clfNames[i] + " classifier")
    plt.legend(loc="lower right")
 
    
    print "Average classification accuracy for %s classifier: %0.4f\n" %(clfNames[i],
                                                                       numpy.mean(cross_val_score(clf,data.iloc[:,firstColLoc:].values,data['GrantStatus'].values,cv=5)))

plt.show()
```
